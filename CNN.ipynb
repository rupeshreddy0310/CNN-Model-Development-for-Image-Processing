---------

1. reshaping----By reshaping the data to (-1, 28, 28), it retains the original height and width of the images (28x28 pixels) while leaving the number of samples unspecified 
(indicated by -1). This allows the reshape operation to automatically determine the number of samples based on the size of the input data.

2. Normalizing---Before feeding the data into the neural network, it's common practice to normalize the pixel values to a range between 0 and 1 by dividing by the

```python
x_train=pd.read_csv('fashion-mnist_train.csv')
x_test=pd.read_csv('fashion-mnist_test.csv')

y_train=x_train.pop('label') #Removes the 'label' column from the training data and assigns it to y_train
y_test=x_test.pop('label') #Removes the 'label' column from the test data and assigns it to y_train

x_train=np.array(x_train).reshape(-1,28,28)/255.0   #increasing the size of the pixel,,, reshaping and normalizing::
x_test=np.array(x_test).reshape(-1,28,28)/255.0
```

```python
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)
 
def make_and_fit():
    inputs = Input(shape=(28, 28, 1))
 
    x = Conv2D(32, (3, 3), activation='relu', padding='same')(inputs)
    x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
    x = MaxPooling2D((2, 2))(x)
    x = BatchNormalization()(x)
    x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
    x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
    x = Conv2D(32, (2, 2), activation='relu', padding='same')(x)
    x = MaxPooling2D((2, 2))(x)
    x = BatchNormalization()(x)
    x = Conv2D(32, (2, 2), activation='relu', padding='same')(x)
    x = Conv2D(32, (2, 2), activation='relu', padding='same')(x)
    x = Conv2D(32, (2, 2), activation='relu', padding='same')(x)
    x = Flatten()(x)
    encoded = Dense(2, activation='relu')(x)
 
    encoder = Model(inputs=inputs, outputs=encoded)
 
    encoded_inputs = Input(shape=(2,))
 
    x = Dense(4, activation='relu')(encoded_inputs)
    x = Reshape((2, 2, 1))(x)
    x = Conv2D(4, (2, 2), activation='relu', padding='same')(x)
    x = Conv2D(16, (2, 2), activation='relu', padding='same')(x)
    x = BatchNormalization()(x)
    x = UpSampling2D(size=(7, 7))(x)  # Fix the typo here
    x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
    x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
    x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
    x = BatchNormalization()(x)
    x = UpSampling2D(size=(2, 2))(x)  # Fix the typo here
    x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
    decoded = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
 
    decoder = Model(inputs=encoded_inputs, outputs=decoded)
 
    x = encoder(inputs)
    x = decoder(x)
    model = Model(inputs=inputs, outputs=x)
    model.compile(optimizer=optimizers.Adam(1e-3), loss='binary_crossentropy', metrics=['accuracy', 'mse'])
 
    print(model.summary())
   
    clr = ReduceLROnPlateau(
       monitor='loss',
       factor=0.5,
       patience=3,
       min_delta=0.01,
       cooldown=0,
       min_lr=1e-7,
       verbose=1)
    model.fit(
        x_train,
        x_train,
        batch_size=256,
        epochs=50,
        shuffle=True,
        validation_data=(x_test, x_test),
        callbacks=[clr])
 
    return model, encoder, decoder
 
model, encoder, decoder = make_and_fit()
```
